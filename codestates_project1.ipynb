```python
def to_K(s):
    if (s.endswith('K')):
        return float(s.replace('K', ''))
    if (s.endswith('M')):
        return float(s.replace('M', '')) * 1000
    else:
        return float(s) * 1000
```

```python
df['NA_Sales'] = df['NA_Sales'].apply(to_K)
df['EU_Sales'] = df['EU_Sales'].apply(to_K)
df['JP_Sales'] = df['JP_Sales'].apply(to_K)
df['Other_Sales'] = df['Other_Sales'].apply(to_K)
```

```python
from numpy import NaN
def to_fourth(i):
    try:
        if (i == NaN):
            return i
        if (i >= 100):
            return int(i)
        if (i > 22):
            return int(i) + 1900
        return int(i) + 2000
    except Exception:
        return i
```

```python
def genre_region(dataframe, region):
    if (region not in ['NA', 'EU', 'JP', 'Other']):
        print('region must be one of [NA, EU, JP, Other]')
        return
    
    global genres

    genre_map = {}
    for g in genres:
        genre_map[g] = []
    period = []

    year = 1980

    c_name = region + '_Sales(K)'

    while(1):
        if (year >= 2020):
            break

        for g in genres:
            genre_map[g].append(0)
        period.append(year)
        
        data = []
        if (year in dataframe.index):
            data = dataframe.loc[year]

        # s = data.loc[:, c_name].sum()

        for j in range(len(data)):
            g = data.iloc[j]['Genre']
            genre_map[g][-1] += data.iloc[j][c_name]

        year += 1

    return pd.DataFrame(genre_map, index = period)
```

```python
def platform_sales(dataframe, region = 'whole', ratio = False):
    c = region + '_Sales(K)'
    if (region == 'whole'):
        c = 'Sales_Sum(K)'
    
    platforms = dataframe['Platform'].unique()

    start = int(dataframe.index.min())
    end = int(dataframe.index.max())

    platform_map = {}
    for p in platforms:
        platform_map[p] = []
    
    period = []
    
    for year in range(start, end + 1):
        
        period.append(year)
        
        for p in platforms:
            platform_map[p].append(1)
        
        data = []
        if (year in dataframe.index):
            data = dataframe.loc[year]

        for j in range(len(data)):
            platform = data.iloc[j]['Platform']
            platform_map[platform][-1] += data.iloc[j][c]
        
    df = pd.DataFrame(platform_map, index = period)
    if (ratio == False):
        return df
    
    df['Sum'] = df.sum(axis = 1)
    for col in df.columns:
        if (col == 'Sum'):
            continue
        df[col] = df[col] / df['Sum']
    return df
```
